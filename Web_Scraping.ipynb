```python
#importing libraries for web scraping

from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
```

```python
#reading the url

httpObject = urlopen("https://www.flipkart.com/q/best-laptops-under-rs-50000")
webdata = httpObject.read()
print(webdata)
```

```python
soup(webdata)
```

```python
soupdata = soup(webdata)
type(soupdata)
```

```python
f = open('Laptops_info.csv','w')
f.write('ProductName,Stars,Ratings,Reviews,CurrentPrice,MRP,Processor,RAM,ImageURL\n')
import re
for container in containers:
  #printing Product names
  product = container.findAll('div',{'class':'_4rR01T'})
  ProductName = product[0].text.split('-')[0].strip()

  #printing stars
  star = container.find('div',{'class':'_3LWZlK'})
  try:
    Stars = star.text
  except:
    Stars = 0


 #printing ratings and reviews
  Rating = container.find('span',{'class':'_2_R_DZ'})
  try:
    ratrev = re.findall('\d+,?\d*',Rating.text)
    Ratings = ratrev[0].replace(',','')
    Reviews = ratrev[1].replace(',','')
  except:
    Ratings = 0
    Reviews = 0


 #printing current price
  CurrentPrice = container.find('div',{'class':'_30jeq3 _1_WHN1'}).text.replace(',','')

 #finding mrp
  mrp = container.find('div',{'class':'_3I9_wc _27UcVY'})
  try:
    MRP=mrp.text.replace(',','')
  except:
    MRP=0


#printing product information
  info = container.findAll('li',{'class':'rgWa7D'})
  Processor = info[0].text
  RAM = info[1].text


#printing image URL
  Image = container.img
  ImageURL = Image.get('src')

  print(ProductName, Stars, Ratings, Reviews, CurrentPrice, MRP, Processor, RAM, ImageURL)
  f.write(f"{ProductName},{Stars},{Ratings},{Reviews},{CurrentPrice},{MRP},{Processor},{RAM},{ImageURL}\n")
  print('\n')
f.close()
```
